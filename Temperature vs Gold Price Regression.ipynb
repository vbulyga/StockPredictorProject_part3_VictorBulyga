```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Create the two arrays we will be using to "train" a linear model
#dailyTemp will serve as the predictor variable while goldVal will be our response variable
dailyTemp = np.array([47.3, 50.3, 50, 49.8, 50.6, 50.7, 50.2, 49.2, 55.6, 52.4,
                      57.5, 57, 57.2, 55.1, 56.7, 53.6, 56.3, 61.1, 55.6, 54.5,
                      49.9, 49, 52.2, 55.5, 53.3, 54.6, 56, 57.7, 50.7, 47.4,
                      52.2]).reshape(-1, 1)

goldVal = np.array([63.32, 63.71, 63.02, 63.23, 63.01, 63.5, 63.62, 63.62, 64.08,
                    63.58, 62.38, 63.78, 63.43, 63.92, 63.92, 61.97, 61.49, 61.17,
                    61.2, 60.54, 60.03, 60.03, 58.83, 58.33, 58.24, 59.4, 59.63,
                    59.65, 59.65, 59.12, 58.99]).reshape(-1, 1)
```

```python
model = LinearRegression()
#This trains the our linear model
#It calculates the slope and y-intercept of the line that best fits our data.
#This calculated line gives predicted response values that have a differ minimally from the true response values.
model.fit(dailyTemp, goldVal)
```

```python
#plots raw data
plt.scatter(dailyTemp, goldVal)
#plots our predicted response values vs raw predictor values 
plt.plot(dailyTemp, model.predict(dailyTemp), color = 'red')
#printing the slope and intercept of our prediction line
#print(<slope>, <intercept>)
print(model.coef_, model.intercept_)

#Changing the X-axis and Y-axis titles
plt.xlabel('Avg Daily temp (F) March 2023 in Los Angeles, CA')
plt.ylabel('Price of Gold ($ per gram) March 2023')

#Calculating and printing the R^2 term which indicates how well our model fits the data
r_squared = model.score(dailyTemp, goldVal)
print("R^2:", r_squared)
```
